```python
import pandas as pd
import gseapy as gp
from gseapy.parser import read_gmt
import os

os.chdir("/slurm/home/yrd/liaolab/nieshuyang/AHBA_sc")
```

```python
fnames = [
    "SVG(+)_L56 IT",
    "SVG(+)_PVALB",
    "SVG(-)_L3-5 IT",
    "SVG(-)_L56 IT Car3"
]
BP = read_gmt("/GMT/GO_Biological_Process_2025.gmt")
CC = read_gmt("/GMT/GO_Cellular_Component_2025.gmt")
MF = read_gmt("/GMT/GO_Molecular_Function_2025.gmt")
GWAS = read_gmt("/GMT/GWAS_Catalog_2025.gmt")
gmt_dict = {
    "BP": BP, 
    "CC": CC, 
    "MF": MF,
    "GWAS": GWAS
}

for fname in fnames:
    gene_rnk = pd.read_csv(f"/HCPMMP/SVG/DEG/{fname}.csv",index_col=0)
    res_dict = {}
    for gmt in gmt_dict.keys():
        gsea_res = gp.prerank(rnk=gene_rnk, gene_sets=gmt_dict[gmt],
                              threads=-1, min_size=50, max_size=300,
                              permutation_num=1000, outdir=None, 
                              seed=123, verbose=False)
        gsea_res.res2d["Gene Set"] = gmt
        res_dict[gmt] = gsea_res.res2d
    res = pd.concat(res_dict.values(), ignore_index=True)
    res = res.set_index("Term")
    res.to_csv(f"/HCPMMP/SVG/GSEA/{fname}.csv")
```
